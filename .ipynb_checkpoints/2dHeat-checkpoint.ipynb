d^2u/dt^2 = f(x) = sin(pi.x)  0<x<1  
u(0) = 0 u(1) = 0

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def func(x):
    return np.sin(np.pi*x)
```

```python
N = 20
h = 1/(N+1)
u_left = 0
u_right = 0

x = np.zeros(N)
u = np.zeros(N)
ue = np.zeros(N)
f = np.zeros(N)
A = np.zeros((N,N))

for i in range(1,N+1):
    x[i-1] = i*h
    f[i-1]= h*h*func(x[i-1])
    ue[i-1] = func(x[i-1])/np.pi/np.pi
    if i>1:
        A[i-1,i-2] = -1
    else:
        f[i-1]=f[i-1] + u_left
    A[i-1,i-1] = 2
    if i<N:
        A[i-1,i] = -1
    else:
        f[i-1] = f[i-1] + u_right
u = np.linalg.solve(A,f)
```

```python
plt.plot(u ,label = "u")
plt.plot(ue ,label = "ue")
plt.legend()
```
